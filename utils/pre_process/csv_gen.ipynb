```python
f = open("CCBD/Project/Colour matching data", "r")
data = f.readlines()

col1 = []
col2 = []
for item in data:
    col1.append(item.split('\t')[0].strip())
    col2.append(item.split('\t')[1].strip())

print(col1)
print(col2)
```

```python
import pandas as pd

d = {'Dress 1': col1, 'Dress 2': col2}
df = pd.DataFrame(d)
df.to_csv('Colour.csv')
df  
```
